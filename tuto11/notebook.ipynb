#### Importation des Bibliothèques

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

#### Graphique 2D - Diagramme à Barres pour la Résistance des Matériaux

```python
# Données
matériaux = ['Béton', 'Acier', 'Bois', 'Aluminium']
résistances = [30, 300, 20, 100]

# Tracé du diagramme à barres
plt.bar(matériaux, résistances, 
        color=['blue', 'gray', 'brown', 'silver'])

# Ajout de titres et d'étiquettes
plt.title('Résistance des Matériaux')
plt.xlabel('Matériaux')
plt.ylabel('Résistance (MPa)')

# Affichage du graphique
plt.show()
```

#### Graphique 2D - Diagramme de dispersion pour les Charges sur une Structure

```python
# Données
charges = [500, 800, 1200, 600, 1000]
longueurs = [5, 8, 12, 6, 10]

# Tracé du diagramme de dispersion
plt.scatter(longueurs, charges, color='red', marker='o')

# Ajout de titres et d'étiquettes
plt.title('Charges sur une Structure')
plt.xlabel('Longueur de la Structure (m)')
plt.ylabel('Charge (kN)')

# Affichage du graphique
plt.show()
```

#### Graphique 3D - Modélisation d'une Poutre en 3D

```python
# Données
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.zeros_like(x)

# Tracé de la poutre en 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, color='green', linewidth=2)

# Ajout de titres et d'étiquettes
ax.set_title('Modélisation d\'une Poutre en 3D')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Affichage du graphique
plt.show()
```

#### Graphique 3D - Diagramme de Surface pour l'Élévation du Terrain

```python
# Données
x = np.linspace(0, 10, 50)
y = np.linspace(0, 10, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Tracé du diagramme de surface
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

# Ajout de titres et d'étiquettes
ax.set_title('Élévation du Terrain en 3D')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Affichage du graphique
plt.show()
```
