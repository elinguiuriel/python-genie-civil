#### Importation des Bibliothèques

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Personnalisation d'un Graphique de Résistance des Matériaux

```python
# Données
matériaux = ['Béton', 'Acier', 'Bois', 'Aluminium']
résistances = [30, 300, 20, 100]

# Tracé du diagramme à barres
plt.bar(matériaux, résistances, color=['blue', 'gray', 'brown', 'silver'])

# Personnalisation du graphique
plt.title('Résistance des Matériaux', fontsize=16)
plt.xlabel('Matériaux', fontsize=12)
plt.ylabel('Résistance (MPa)', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# Ajout d'une grille
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Ajout d'une légende
plt.legend(['Résistance'], loc='upper right')

# Affichage du graphique
plt.show()
```

#### Personnalisation d'un Diagramme de Dispersion pour les Charges sur une Structure

```python
# Données
charges = [500, 800, 1200, 600, 1000]
longueurs = [5, 8, 12, 6, 10]

# Tracé du diagramme de dispersion
plt.scatter(longueurs, charges, color='red', marker='o')

# Personnalisation du graphique
plt.title('Charges sur une Structure', fontsize=16)
plt.xlabel('Longueur de la Structure (m)', fontsize=12)
plt.ylabel('Charge (kN)', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# Ajout d'une grille
plt.grid(linestyle='--', alpha=0.7)

# Affichage du graphique
plt.show()
```

#### Personnalisation d'un Graphique en 3D - Modélisation d'une Poutre

```python
# Données
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.zeros_like(x)

# Tracé de la poutre en 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, color='green', linewidth=2)

# Personnalisation du graphique en 3D
ax.set_title('Modélisation d\'une Poutre en 3D', fontsize=16)
ax.set_xlabel('X', fontsize=12)
ax.set_ylabel('Y', fontsize=12)
ax.set_zlabel('Z', fontsize=12)

# Affichage du graphique
plt.show()
```

#### Personnalisation d'un Graphique en 3D - Diagramme de Surface pour l'Élévation du Terrain

```python
# Données
x = np.linspace(0, 10, 50)
y = np.linspace(0, 10, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Tracé du diagramme de surface
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='k', linewidth=0.5)

# Personnalisation du diagramme de surface
ax.set_title('Élévation du Terrain en 3D', fontsize=16)
ax.set_xlabel('X', fontsize=12)
ax.set_ylabel('Y', fontsize=12)
ax.set_zlabel('Z', fontsize=12)
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=10) # Ajout d'une barre de couleur

# Affichage du graphique
plt.show()
```
